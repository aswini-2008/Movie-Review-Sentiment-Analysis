```python
# ==========================================
# MOVIE REVIEW SENTIMENT ANALYSIS
# Growfinix Internship Task 4
# ==========================================

# STEP 1 - Import Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from textblob import TextBlob

# STEP 2 - Create Dataset

data = {
    "Review": [
        "This movie was amazing and fantastic",
        "I loved the acting and story",
        "Worst movie I have ever seen",
        "The film was boring and slow",
        "Excellent direction and great music",
        "Terrible movie and poor acting",
        "I really enjoyed this movie",
        "The movie was not good",
        "Outstanding performance by actors",
        "Waste of time watching this film"
    ]
}

df = pd.DataFrame(data)

print("Dataset:\n")
print(df)

# STEP 3 - Function for Sentiment Analysis

def get_sentiment(text):

    analysis = TextBlob(text)

    polarity = analysis.sentiment.polarity

    if polarity > 0:
        return "Positive"

    elif polarity < 0:
        return "Negative"

    else:
        return "Neutral"

# STEP 4 - Apply Sentiment Analysis

df["Sentiment"] = df["Review"].apply(get_sentiment)

print("\nSentiment Results:\n")
print(df)

# STEP 5 - Count Sentiments

sentiment_count = df["Sentiment"].value_counts()

print("\nSentiment Counts:\n")
print(sentiment_count)

# STEP 6 - Bar Chart

plt.figure(figsize=(6,5))

sns.countplot(
    x="Sentiment",
    data=df
)

plt.title("Movie Review Sentiment Analysis")

plt.savefig("sentiment_bar_chart.png")

plt.show()

# STEP 7 - Pie Chart

plt.figure(figsize=(6,5))

plt.pie(
    sentiment_count,
    labels=sentiment_count.index,
    autopct='%1.1f%%'
)

plt.title("Sentiment Distribution")

plt.savefig("sentiment_pie_chart.png")

plt.show()

# STEP 8 - Save Dataset

df.to_csv(
    "movie_sentiment_results.csv",
    index=False
)

print("\nDataset Saved Successfully!")

# STEP 9 - Conclusion

print("\n========== CONCLUSION ==========")

print("1. Positive reviews are identified correctly.")

print("2. Negative reviews are classified successfully.")

print("3. Sentiment analysis helps understand customer opinions.")
```
